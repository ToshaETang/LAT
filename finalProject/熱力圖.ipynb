```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
font = {'family': 'MicroSoft YaHei'}
plt.rc('font',**font)
#讀取大學資料
df1=pd.read_csv('中原大學.csv')
df2=pd.read_csv("中國醫藥大學.csv")
df3=pd.read_csv('中山醫學大學.csv')
df4=pd.read_csv('中華大學.csv')
df5=pd.read_csv('元智大學.csv')
df6=pd.read_csv('南華大學.csv')
df7=pd.read_csv("台北基督學院.csv")
df8=pd.read_csv("台北醫學大學.csv")
df9=pd.read_csv("國立台南大學.csv")
df10=pd.read_csv("國立台灣海洋大學.csv")
df11=pd.read_csv("國立嘉義大學.csv")
df12=pd.read_csv("國立屏東大學.csv")
df13=pd.read_csv("國立聯合大學.csv")
df14=pd.read_csv("國立臺灣師範大學.csv")
df15=pd.read_csv("康寧大學.csv")
df16=pd.read_csv("明道大學.csv")
df17=pd.read_csv("東吳大學.csv")
df18=pd.read_csv("玄奘大學.csv")
df19=pd.read_csv("逢甲大學.csv")
df20 = pd.read_csv("銘傳大學.csv")
df21 = pd.read_csv('長庚大學.csv')
df22 = pd.read_csv('長榮大學.csv')
df23 = pd.read_csv('靜宜大學.csv')
# 計算科目比例
def calculate_ratios(data):
    國count = (data['採計科目(含一階篩選方式、倍率、超篩)'].str.contains('國')).sum()
    國countratio = 國count / len(data)

    英count = (data['採計科目(含一階篩選方式、倍率、超篩)'].str.contains('英')).sum()
    英countratio = 英count / len(data)

    數Bcount = (data['採計科目(含一階篩選方式、倍率、超篩)'].str.contains('B')).sum()
    數Bcountratio = 數Bcount / len(data)

    數Acount = (data['採計科目(含一階篩選方式、倍率、超篩)'].str.contains('A')).sum()
    數Acountratio = 數Acount / len(data)

    社count = (data['採計科目(含一階篩選方式、倍率、超篩)'].str.contains('社')).sum()
    社countratio = 社count / len(data)

    自count = (data['採計科目(含一階篩選方式、倍率、超篩)'].str.contains('自')).sum()
    自countratio = 自count / len(data)

    英聽count = (data['採計科目(含一階篩選方式、倍率、超篩)'].str.contains('聽')).sum()
    英聽countratio = 英聽count / len(data)

    return [國countratio, 英countratio, 數Acountratio, 數Bcountratio, 社countratio, 自countratio, 英聽countratio]

# 計算兩個學校的科目比例
ratios_df1 = calculate_ratios(df1)
ratios_df2 = calculate_ratios(df2)
ratios_df3 = calculate_ratios(df3)
ratios_df4 = calculate_ratios(df4)
ratios_df5 = calculate_ratios(df5)
ratios_df6 = calculate_ratios(df6)
ratios_df7 = calculate_ratios(df7)
ratios_df8 = calculate_ratios(df8)
ratios_df9 = calculate_ratios(df9)
ratios_df10 = calculate_ratios(df10)
ratios_df11 = calculate_ratios(df11)
ratios_df12 = calculate_ratios(df12)
ratios_df13 = calculate_ratios(df13)
ratios_df14 = calculate_ratios(df14)
ratios_df15 = calculate_ratios(df15)
ratios_df16 = calculate_ratios(df16)
ratios_df17 = calculate_ratios(df17)
ratios_df18 = calculate_ratios(df18)
ratios_df19 = calculate_ratios(df19)
ratios_df20 = calculate_ratios(df20)
ratios_df21 = calculate_ratios(df21)
ratios_df22= calculate_ratios(df22)
ratios_df23= calculate_ratios(df23)
# 將科目比例存儲在二維列表中
科目比例 = [ratios_df1,ratios_df2,ratios_df3,ratios_df4,ratios_df5,ratios_df6,ratios_df7,ratios_df8,ratios_df9,ratios_df10,ratios_df11,ratios_df12,ratios_df13,ratios_df14,ratios_df15,ratios_df16,ratios_df17,ratios_df18,ratios_df19,ratios_df20, ratios_df21,ratios_df22,ratios_df23]

# 設置大學名稱列表
大學名稱 = ['中原大學', '中國醫藥大學',"中山醫學大學","中華大學","元智大學","南華大學","台北基督學院","台北醫學大學","國立台南大學",
        "國立台灣海洋大學","國立嘉義大學","國立屏東大學","國立聯合大學","國立臺灣師範大學","康寧大學","明道大學","東吳大學",
        "玄奘大學","逢甲大學","銘傳大學","長庚大學","長榮大學","靜宜大學"] 
plt.figure(figsize=(8, 10))
# 繪製熱力圖
sns.heatmap(科目比例, cmap='coolwarm', annot=True, fmt=".2f")
plt.xticks(rotation=45)  # 旋轉橫軸刻度標籤
plt.yticks(rotation=45)  # 旋轉縱軸刻度標籤
# 設置刻度標籤和軸標籤
plt.xticks(range(len(科目比例[0])), ['國文', '英文', '數A', '數B', '社會', '自然', '英聽'])
plt.yticks(range(len(大學名稱)), 大學名稱)

# 設置圖表標題和軸標籤等
plt.title("各科目比例熱力圖")
plt.xlabel("科目")
plt.ylabel("大學名稱")

# 顯示圖表
plt.show()
```
